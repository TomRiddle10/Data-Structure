```python
queue = []

print("1. Enqueue")
print("2. Dequeue")
print("3. Show queue")
print("4. Remove by index")
print("5. Exit")

while True:
    choice = int(input("Enter number for performing operation (Within range of 1-5): "))

    if choice == 1:
        value = input("Enter a value: ")
        queue.append(value)
        print(f"Queue after enqueue: {queue}")

    elif choice == 2:
        if queue:
            removed = queue.pop(0)
            print(f"Dequeued value: {removed}")
        else:
            print("Queue is empty. Nothing to dequeue.")
        print(f"Queue after dequeue: {queue}")

    elif choice == 3:
        print(f"Current queue: {queue}")

    elif choice == 4:
        if queue:
        print(f"Current queue: {queue}")
            try:
            index = int(input("Enter index to remove (starting from 1): ")) - 1  # Adjust for 0-based indexing
            if 0 <= index < len(queue):
                removed = queue.pop(index)
                print(f"Removed value at position {index + 1}: {removed}")
            else:
                print("Invalid index. Please enter a valid index within range.")
        except ValueError:
            print("Invalid input. Please enter a numeric index.")
        print(f"Queue after index removal: {queue}")
    else:
        print("Queue is empty. Nothing to remove.")

    elif choice == 5:
        print("Exiting program.")
        break

    else:
        print("Invalid input. Please enter a number between 1 and 5.")
```
