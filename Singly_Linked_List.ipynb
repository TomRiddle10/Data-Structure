```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, value):
        new_node = Node(value)
        if not self.head:
            self.head = new_node
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node
        print(f"Inserted '{value}' at the end of the list.")

    def delete_by_index(self, index):
        if not self.head:
            print("List is empty. Nothing to delete.")
            return

        index -= 1  # Adjust for 0-based indexing
        temp = self.head

        if index == 0:
            removed = temp.data
            self.head = temp.next
            print(f"Deleted node at position 1: {removed}")
            return

        prev = None
        count = 0
        while temp and count < index:
            prev = temp
            temp = temp.next
            count += 1

        if temp:
            removed = temp.data
            prev.next = temp.next
            print(f"Deleted node at position {index + 1}: {removed}")
        else:
            print("Invalid index. No node found at that position.")

    def delete_last(self):
        if not self.head:
            print("List is empty. Nothing to delete.")
            return

        if not self.head.next:
            removed = self.head.data
            self.head = None
            print(f"Deleted last node: {removed}")
            return

        temp = self.head
        while temp.next and temp.next.next:
            temp = temp.next

        removed = temp.next.data
        temp.next = None
        print(f"Deleted last node: {removed}")

    def show(self):
        if not self.head:
            print("Linked list is empty.")
            return
        temp = self.head
        print("Linked list contents:")
        pos = 1
        while temp:
            print(f"{pos}. {temp.data}")
            temp = temp.next
            pos += 1
```

```python
ll = LinkedList()

print("1. Insert at end")
print("2. Delete by index (starting from 1)")
print("3. Delete last element")
print("4. Show linked list")
print("5. Exit")

while True:
    choice = int(input("\nEnter your choice (1-Insert, 2-Delete  element by index, 3-Delete last element, 4-Show, 5-Exit): "))

    if choice == 1:
        value = input("Enter value to insert: ")
        ll.append(value)
        ll.show()

    elif choice == 2:
        try:
            index = int(input("Enter position to delete element(starting from 1): "))
            ll.delete_by_index(index)
            ll.show()
        except ValueError:
            print("Invalid input. Please enter a numeric index.")

    elif choice == 3:
        ll.delete_last()
        ll.show()

    elif choice == 4:
        ll.show()

    elif choice == 5:
        print("Exiting program.")
        break

    else:
        print("Invalid choice. Please enter a number between 1 and 5.")
```
